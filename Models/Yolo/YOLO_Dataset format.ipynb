Converting all the images to jpg format

```python
for i in os.listdir("Data"):
    for images in os.listdir(f"Data/{i}/images"):
        name = images.split(".")[0]
        format = images.split(".")[1]
        if format != "jpg":
            os.rename(f"Data/{i}/images/{images}", f"Data/{i}/images/{name}.jpg")
```

Converting the annotations

```python
class_names = os.listdir("ExDark")
class_dict = dict([(class_names[i], i) for i in range(len(class_names))])
class_dict
```

```python
for dir in os.listdir("datasets"):
    for annot in os.listdir(f"datasets/{dir}/labels"):
        file = f"Data/{dir}/labels/{annot}"
        raw_file = open(file, "r")
        raw_cont = raw_file.read().strip().split("\n")[1:]
        new_cont = ""
        for i in raw_cont:
            obj_list = i.split(" ")[:5]
            obj_list[0] = class_dict[obj_list[0]]

            # Normalizing part
            bbox = np.array(obj_list[1:]).astype(float)
            bbox = np.round([float(i)/np.sum(bbox) for i in bbox], 3)
            
            obj_list[1:] = bbox
            new_cont += " ".join(map(str, obj_list)) + "\n"
        print(new_cont, file)
        print()

        with open(file, "w") as file:
            file.write(new_cont)
```

```python
import matplotlib.pyplot as plt


for dir in os.listdir("datasets"):
    for annot in os.listdir(f"datasets/{dir}/labels"):
        file = f"datasets/{dir}/labels/{annot}"
        image_file = f"datasets/{dir}/images/{annot.split('.')[0]}.jpg"
        image = plt.imread(image_file)

        raw_file = open(file, "r")
        raw_cont = raw_file.read().strip().split("\n")[1:]
        new_cont = ""
        for i in raw_cont:
            obj_list = i.split(" ")[:5]
            obj_list[0] = class_dict[obj_list[0]]

            img_h, img_w = image.shape[:2]

            x, y, w, h = np.array(obj_list[1:]).astype(float)
            x_center = ((x + w) / 2) / img_w
            y_center = ((y + h) / 2) / img_h
            width = w / img_w
            height = h / img_h
            # print(x, y, w, h)
            # print(x_center, y_center, width, height)

            obj_list[1:] = x_center, y_center, width, height
            new_cont += " ".join(map(str, obj_list)) + "\n"

        with open(file, "w") as file:
            file.write(new_cont)
```
